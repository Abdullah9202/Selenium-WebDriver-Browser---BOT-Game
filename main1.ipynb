# Selecting and finding elements on website using selenium

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

AMZ_URL = "https://www.amazon.com/LG-27UL550-W-Monitor-Freesync-Technology/dp/B07PX8JMJV/ref=sr_1_2?crid=2TH83ST0BTST8&keywords=4k%2Bled%2Bmonitor&qid=1702799316&sprefix=4k%2Bled%2Bmonitor%2Caps%2C296&sr=8-2&th=1"

# Setting Up Chrome Options
chrome_options = webdriver.ChromeOptions()
# Setup the chrome driver
chrome_Driver = webdriver.Chrome(options=chrome_options)
# Getting the data from URL
chrome_Driver.get(url=AMZ_URL)
# Scrapping the data
price = chrome_Driver.find_element(By.CSS_SELECTOR, value="span.a-price-whole")
print(f"${price.text}")
# Closing
chrome_Driver.close()
```

# Finding elements by name, classes and Css Selectors

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

PY_URL = "https://www.python.org/"

# Setting up chrome option
chrome_options = webdriver.ChromeOptions()
# Setting up chrome drivers
chrome_Driver = webdriver.Chrome(options=chrome_options)
chrome_Driver.get(url=PY_URL)
# Finding elements by names
py_Search_Bar = chrome_Driver.find_element(By.NAME, value="q")
print(py_Search_Bar.tag_name, py_Search_Bar.get_attribute("placeholder"))
# Finding elements by classes
py_Logo = chrome_Driver.find_element(By.CLASS_NAME, value="python-logo")
print(py_Logo.size)
# Finding elements by css selectors
py_Docs = chrome_Driver.find_element(By.CSS_SELECTOR, value="div.documentation-widget p a")
print(py_Docs.text)
# Closing
chrome_Driver.close()
```

# Finding elements by XPath

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

PY_URL = "https://www.python.org/"
# Setting up chrome option
chrome_options = webdriver.ChromeOptions()
# Setting up driver
chrome_Driver = webdriver.Chrome(options=chrome_options)
chrome_Driver.get(url=PY_URL)
# Finding elements by Xpath
web_Bug_Link = chrome_Driver.find_element(By.XPATH, value='//*[@id="site-map"]/div[2]/div/ul/li[3]/a')
print(f"{web_Bug_Link.text} on {web_Bug_Link.get_attribute('href')}")
# Closing
chrome_Driver.close()
```

# Finding multiple elements using find_elements()

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

PY_URL = "https://www.python.org/"
# Setting up chrome option
chrome_options = webdriver.ChromeOptions()
# Setting up driver
chrome_Driver = webdriver.Chrome(options=chrome_options)
chrome_Driver.get(url=PY_URL)
# Finding multiple elements using find_elements()
mult_elements = chrome_Driver.find_elements(By.TAG_NAME, value="a")
for link in mult_elements:
    print(link.get_attribute('href'))
    
# Closing
chrome_Driver.close()
```
