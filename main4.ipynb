# Cookie clicker (Experiment) using selenium

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

def clickCookie():
    cookie_Button.click()
    cookie_Button.click()
    cookie_Button.click()
    cookie_Button.click()
    cookie_Button.click()

CCE_URL = "https://orteil.dashnet.org/experiments/cookie/"
# firefoxOptions
firefox_options = webdriver.FirefoxOptions()
# ChromeDriver
firefox_Driver = webdriver.Firefox(options=firefox_options)
# Maximizing the screen
firefox_Driver.maximize_window()
# Sending the request
firefox_Driver.get(url=CCE_URL)
# Finding the cookie button
cookie_Button = firefox_Driver.find_element(By.ID, value="cookie")

check_after_5sec = time.time() + 5 # 5 Seconds ahead of current time
timeout = time.time() + 60*5 # 300 seconds = 5 min

while True:
    # Clicking the cookie
    clickCookie()
    item_Price = [] # List to store the int prices
    facility_Dict = {} # To store the price and facility
    # Checking after every 5 seconds
    if time.time() >= check_after_5sec:
        store_Items = firefox_Driver.find_elements(By.CSS_SELECTOR, value="div#rightPanel div#store b") # List to store the items and prices
        # Converting the prices into integers
        for item in store_Items:
            if item.text != "": 
                price = item.text.split("-")[1].strip().replace(",", "")
                # Appending in list
                item_Price.append(price)
                
        # Getting the ids of facilities
        items = firefox_Driver.find_elements(By.CSS_SELECTOR, value="div.rightPanel div#store div")
        item_ids = [item.get_attribute("id") for item in items]
        # Appening in dict
        # for x in range(len(item_Price)):
        #     if x < len(item_Price): # Cheking if the x within the bounds
        #         facility_Dict[item_Price[x]] = item_ids[x]
        facility_Dict = {Price : id for (Price, id) in zip(item_Price, item_ids)}

        # Getting the current cookie count (Money)
        money_element = firefox_Driver.find_element(By.ID, value="money").text
        # Removing the "," if found
        if "," in money_element:
            money_element = money_element.replace(",", "")
        money = int(money_element) # int money
        
        affordable_Upgrades = {} # Dict to store the affordable upgrades
        # Comparing the money with buyable items
        for cost, id in facility_Dict.items():
            if money >= cost:
                affordable_Upgrades[cost] = id # Appending the afforadble items into dict
        
        # Purchasing the most expensive upgrade
        to_Purchase_ID = "" # To store the ID of buyable upgrade
        if affordable_Upgrades:
            most_Expensive_Affordable_Upgrade = max(affordable_Upgrades)
            print(f"Most expensive and affordable upgrade: {most_Expensive_Affordable_Upgrade}")
            to_Purchase_ID = affordable_Upgrades[most_Expensive_Affordable_Upgrade]
        
        # Purchasing the most affordable upgrade
        if to_Purchase_ID:
            print(f"To purchase the upgrade: {to_Purchase_ID}")
            firefox_Driver.find_element(By.ID, value=to_Purchase_ID).click()
        
        # Adding another 5 seconds until next check
        check_after_5sec = time.time() + 5
        
    # Stopping the BOT after 5 seconds and checking the cookies
    if time.time() >= timeout:
        cps = firefox_Driver.find_element(By.ID, value="cps").text
        print(f"Cookies/second: {cps}")
        break

    
    
        
```
